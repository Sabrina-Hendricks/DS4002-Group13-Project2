```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Import data from Github

```python
url = 'https://raw.githubusercontent.com/Sabrina-Hendricks/DS4002-Group13-Project2/refs/heads/main/Data/HealthCareOriginalData.csv'
df = pd.read_csv(url)
df.head()
```

```python
df = df.set_index('Unnamed: 0')
df.head()
```

```python
# Step 1: Reshape the data by using melt, categorizing by year, quarter, and spending
df_melted = df.melt(id_vars=['Unnamed: 0'], var_name='year_quarter', value_name='spending')

# Step 2: Split the 'year_quarter' into separate year and quarter columns
df_melted['year'] = df_melted['year_quarter'].apply(lambda x: x.split('.')[0])
df_melted['quarter'] = df_melted['year_quarter'].apply(lambda x: 'Q' + x.split('.')[1] if '.' in x else 'Q1')

# Step 3: Create category column for hierarchy - categorize based on Healthcare sections
df_melted['category'] = df_melted['Unnamed: 0'].apply(lambda x: 'Healthcare' if 'Health care' in x else
                                                    'Outpatient services' if 'Outpatient' in x or 'Physician' in x or 'Dental' in x or 'Paramedical' in x else
                                                    'Hospital and nursing home services' if 'Hospital' in x or 'Nursing' in x else x)

# Step 4: Remove unnecessary columns and rename as needed
df_final = df_melted[['category', 'Unnamed: 0', 'year', 'quarter', 'spending']]
df_final.rename(columns={'Unnamed: 0': 'subcategory'}, inplace=True)

# Now df_final will have a readable format with the category, subcategory, year, quarter, and spending
df_final.head()
```
